# Emissions Itinerary

The goal of this project is to help air travelers begin taking carbon emissions into consideration when planning their trips. Using [Climatiq's](https://climatiq.io/) emission factor API and an interactive streamlit app, I plan to let users explore a variety of route options and the associated carbon impact of each. So far, I set up an automated API call loop that collected emissions data for 180,000 international and domestic routes. That data is being stored in MongoDB, and pulled into a dataframe in my streamlit app script. 

The current state of the app is an interactive route map with tooltips showing emissions, which can be filtered by origin & destination airport. 

My next steps for the app include:
- More detailed filtering by origin city/country and destination city/country
- Annotations of how a flight compares to others with similar destinations
- Ability to see only domestic or only international flights 
- Text & explainers on why this app is important
